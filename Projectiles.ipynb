```python
# projectiles 

from math import *
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np
from numpy.random import seed
from numpy.random import rand



```

```python
GRAVITY = -9.81

# Plots the motion of a single projectile with the given
# initial velocity v0, angle, and number of steps to plot.
xx = []
yy = []
def projectile(v0, angle, steps, color="black"):

    # v0y = v0 sin theta, v0x = v0 cos theta
    v0y = v0 * sin(radians(angle))
    v0x = v0 * cos(radians(angle))

    # t = -2v0 / a
    totalTime = -2 * v0y / GRAVITY
    dt = totalTime / steps

    for i in range(steps + 1):
        t = i * dt
        x = v0x * t
        y = displacement(v0y, GRAVITY, t)
        xx.append(x)
        yy.append(y)


# Computes and returns the change in position of a projectile
# given its initial velocity v0, acceleration a, and time t.
def displacement(v0, a, t):
    return v0 * t + 0.5 * a * t**2

```

```python
# Main
projectile(100,45, 100)
print(yy)
fig, axs = plt.subplots(1, 1, figsize=(9, 3), sharey=True)
axs.plot(xx, yy)
```

```python

axs.set_xlim(0,1000)
axs.set_ylim(0,200)
line, = axs.plot(0,0)
def animation_frame(i):
    line.set_xdata(xx[i])
    line.set_ydata(yy[i])
    return line,

animation = FuncAnimation(fig,func=animation_frame, frames = np.arange(0,10,0.1), interval = 10)
plt.show()
```
